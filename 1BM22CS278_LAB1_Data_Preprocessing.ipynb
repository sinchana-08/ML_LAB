```python
df1 = pd.read_csv('/content/adult.csv')
print(df1)
```

```python
df2 = pd.read_csv('/content/Dataset of Diabetes .csv')
print(df2)
```

```python
   # For 'adult.csv' (df1)
missing_cols_adult = df1.columns[df1.isnull().sum() > 0]
print(f"Columns with missing values in 'adult.csv': {missing_cols_adult.tolist()}")

   # For 'Dataset of Diabetes .csv' (df2)
missing_cols_diabetes = df2.columns[df2.isnull().sum() > 0]
print(f"Columns with missing values in 'Dataset of Diabetes .csv': {missing_cols_diabetes.tolist()}")
```

```python
# For 'adult.csv' (df1)
categorical_cols_adult = df1.select_dtypes(include=['object']).columns
print(f"Categorical columns in 'adult.csv': {categorical_cols_adult.tolist()}")

   # For 'Dataset of Diabetes .csv' (df2)
categorical_cols_diabetes = df2.select_dtypes(include=['object']).columns
print(f"Categorical columns in 'Dataset of Diabetes .csv': {categorical_cols_diabetes.tolist()}")
```
